### calculate $V_{\pi}$ in gridworld 5x5

Example 3.5

```python
import numpy as np

nx,ny=5,5
A,A_=[0,1],[4,1]
B,B_=[0,3],[2,3]
actions=[[0,-1],[-1,0],[0,1],[1,0]]#left,up,right,down
pi=0.25
gm=0.9

def step(s,a):
    if s==A:
        return A_,10
    if s==B:
        return B_,5
    
    s_=[s[0]+a[0],s[1]+a[1]]
    if s_[0]<0 or s_[0]>=nx or s_[1]<0 or s_[1]>=ny:
        r=-1
        s_=s
    else:
        r=0
        
    return s_,r

V=np.zeros((nx,ny))

while True:
    V_=np.zeros_like(V)
    for x in range(nx):
        for y in range(ny):
            va=[]
            for a in actions:
                (x_,y_),r=step([x,y],a)
                va.append(pi*(r+gm*V[x_,y_]))
            V_[x,y]=np.sum(va)

    #print(np.around(V,decimals=2))
    if np.sum(np.abs(V-V_))<1e-4:
        print(np.around(V,decimals=1))
        break
    
    V=V_
```

### calculate $V_*$ in gridworld 5x5

Example 3.8

```python
V=np.zeros((nx,ny))

while True:
    V_=np.zeros_like(V)
    for x in range(nx):
        for y in range(ny):
            va=[]
            for a in actions:
                (x_,y_),r=step([x,y],a)
                va.append(r+gm*V[x_,y_])
            V_[x,y]=np.max(va)

    #print(np.around(V,decimals=2))
    if np.sum(np.abs(V-V_))<1e-4:
        print(np.around(V,decimals=1))
        break
    
    V=V_
```

### calculate $V_k$ in gridworld 4x4

4.1 Policy Evaluation - Example 4.1

```python
import numpy as np
import matplotlib.pyplot as plt

nx,ny=4,4
#left, up, right, down
actions=[[0,-1],[-1,0],[0,1],[1,0]] 
pi=0.25
gm=1

def is_terminal(s):
    return s==[0,0] or s==[nx-1,ny-1]

def step(s,a):
    if is_terminal(s):
        return s,0
    
    s_=[s[0]+a[0],s[1]+a[1]]
    
    if s_[0]<0 or s_[0]>=nx or s_[1]<0 or s_[1]>=ny:
        s_=s
        
    r=-1
    return s_,r

def fig_ini(i,iters):
    if i in iters:
        fig=plt.figure(figsize=(6,6))
        ax=fig.add_subplot(1,1,1)
        ax.set_aspect('equal', adjustable='box')
        ax.set_xticks(np.arange(0,nx+1,1))
        ax.set_yticks(np.arange(0,ny+1,1))
        plt.grid()
        plt.ylim((0,ny))
        plt.xlim((0,nx))
        
def fig_close(i,iters):
    
    plt.show()
    plt.clf()
    plt.close()
    
def plot_arrow(i,iters,v_a,x,y):
    scale=0.3
    x0=nx-0.5
    if i in iters:
        if [x,y]==[0,0] or [x,y]==[nx-1,ny-1]:
            pass
        else:  
            v_a=np.around(v_a,decimals=2)
            label=(np.argwhere(v_a == np.max(v_a)))
            if [1] in label: #up
                plt.arrow(y+0.5,x0-x,0,scale,width=0.1, head_width=0.2, head_length=0.1,fc='g', ec='g')
            if [3] in label: #down
                plt.arrow(y+0.5,x0-x,0,-scale,width=0.1, head_width=0.2, head_length=0.1,fc='c', ec='c')
            if [2] in label: #right
                plt.arrow(y+0.5,x0-x,scale,0,width=0.1, head_width=0.2, head_length=0.1,fc='b', ec='b')
            if [0] in label: #left
                plt.arrow(y+0.5,x0-x,-scale,0,width=0.1, head_width=0.2, head_length=0.1,fc='r', ec='r')
            plt.title('iteration:'+str(i))


V=np.zeros((nx,ny))

#selected iterations
iters=[0,1,2,3,10,217]

for i in range(218):
    V_=np.zeros_like(V)
    
    fig_ini(i,iters)
    for x in range(nx):
        for y in range(ny):
            v_a=[]
            for a in actions:
                (x_,y_),r=step([x,y],a)
                v_a.append(pi*(r+gm*V[x_,y_]))
            V_[x,y]=np.sum(v_a)    
            plot_arrow(i,iters,v_a,x,y)
        
    fig_close(i,iters)
    
    if i in iters:
        print('i =',str(i))
        print('V = \n',np.around(V,decimals=1))
    
    
    
    #convergence condition
    if np.sum(np.abs(V-V_))<1e-4:
        print('i =',str(i))
        #print('V = \n',np.around(V,decimals=1))
        break
        
    V=V_
    
```
